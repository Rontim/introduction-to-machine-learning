## Classification and regression

The two major problems in supervised ml are <i>classification</i> and <i>regression</i>

In <b>Classification </b>, the goal is to predict a <i>class label</i>, which is a choice from a predefined lis of possibilities. Classification is sometimes separated into <i>binary classification</i>, which is the special case of distinguishing between exactly two classes, and <i>multiclass classification</i>, which is classification between more than two classes. 

For <b>Regression</b>, the goal is to predict a <i>continuous number</i>, or a <i>floating-point number</i> in programming terms (or real number in mathematical terms). Predicting a personâ€™s annual income from their education, their age, and where they live is anexample of a regression task.

The easiest way to distinguish between classification and regression task is to ask whether there is some kind of continuity in the output. If so, then the task is a regression task.

## Generalization, Overfitting and Undefitting

In sepervised learning our goal is to build a model on a training data and then be able to make accurate prediction on a new, unseen data that has characteristics with the training data.

If a model is able to make accurate prediction on an unseen data set we say it is able to <i>generalize</i> from the training set to the test set. We want to build a model that is able to generalize as accurately as possible.

Building a model that is too complex for the amount of dat we have is called <i>overfitting</i>.It occurs when you fit a model too closely to the particularities of the training set and obtain a model that works well on the training set but is not able to generalize to new data.

On the other hand, if your model is too simple then you might not be able to capture all the aspects of and variability in the data, and your model will do badly even on the training set.This is called <i>underfitting</i>

There is a sweet spot in between that will yield the best generalization performance.
This is the model we want to find.
